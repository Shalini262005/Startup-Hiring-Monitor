```python
company_group = df.groupby("company_name")
```

```python
summary = company_group.agg(
    total_roles=("job_title", "count"),
    remote_roles=("location", lambda x: x.str.contains("Remote", case=False, na=False).sum()),
    unique_departments=("department", "nunique"),
    product_roles=("job_title", lambda x: x.str.lower().str.contains("product").sum()),
    pm_roles=("job_title", lambda x: x.str.lower().str.contains("manager").sum()),
)
```

```python
dept_counts = df.groupby(["company_name", "department"]).size().unstack(fill_value=0)
summary = summary.join(dept_counts, how="left")
```

```python
def assign_tags(row):
    tags = []
    if row.get("Engineering", 0) >= 5:
        tags.append("Engineering Push")
    if row.get("Sales", 0) >= 3:
        tags.append("GTM Team Expansion")
    if row["remote_roles"] == row["total_roles"] and row["total_roles"] > 0:
        tags.append("Remote Hiring Focus")
    if row["product_roles"] > 0 and row["pm_roles"] > 0:
        tags.append("New Product Buildout")
    if row["unique_departments"] >= 4:
        tags.append("Company-Wide Scaling")
    return ", ".join(tags) if tags else "None"
```

```python
summary["growth_tags"] = summary.apply(assign_tags, axis=1)
```
